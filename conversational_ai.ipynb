```python
import spacy
from medspacy.ner import TargetRule
from google.colab import files  # For uploading files in Colab

# Load the MedSpaCy pipeline
nlp = spacy.blank("en")  # Or use a model like "en_core_web_sm"

# Add the TargetMatcher to the pipeline by its existing factory name
if "medspacy_target_matcher" not in nlp.pipe_names:
    nlp.add_pipe("medspacy_target_matcher", last=True)

# Verify that the TargetMatcher is in the pipeline
print("Pipeline components:", nlp.pipe_names)

# Access the TargetMatcher to add rules
target_matcher = nlp.get_pipe("medspacy_target_matcher")

# Add custom rules to recognize entities
target_matcher.add(TargetRule("history of hypertension", "MEDICAL_HISTORY"))
target_matcher.add(TargetRule("allergic to penicillin", "ALLERGY"))
target_matcher.add(TargetRule("diagnosed with diabetes", "DIAGNOSIS"))
target_matcher.add(TargetRule("Test", "Test"))

# Function to process text from an uploaded file
def process_uploaded_file():
    # Upload file
    print("Please upload your text file:")
    uploaded = files.upload()  # This will prompt the user to upload a file

    # Get the uploaded file name
    file_name = list(uploaded.keys())[0]

    # Read the file content
    with open(file_name, "r") as file:
        text = file.read()

    # Process the text with the MedSpaCy pipeline
    doc = nlp(text)

    # Extract and print the recognized entities
    print("\nExtracted Entities:")
    for ent in doc.ents:
        print(f"Entity: {ent.text}, Label: {ent.label_}")

# Call the function to process the uploaded file
process_uploaded_file()
```
